```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train_norm, y_train)

pd.DataFrame({"importance":abs(model.coef_)/abs(model.coef_).sum(),
             "values": model.coef_}, 
             index = X_train.columns).sort_values(by = "importance", ascending = False)
```

# Gradiente Descendiente

Usaremos 1 valor solamente para predecir y ver cÃ³mo va aprendiendo el modelo por cada iteraciÃ³n

```python
x = (diamonds["x"] - diamonds["x"].mean()) / diamonds["x"].std()
y = diamonds["price"]

model_simple = LinearRegression(fit_intercept=True)
model_simple.fit(x.values.reshape(-1,1),y)
print("Valor de a: ", model_simple.coef_[0])
print("Valor de b: ", model_simple.predict(np.zeros((1,1)))[0])
```

```python
a0 = 1
b0 = 1
alpha = 0.1
grad_a = -2*np.mean( (y - a0*x - b0)*x  )
grad_b = -2*np.mean(y - a0*x - b0)
aes = [a0]
bes = [b0]

grads_a = [grad_a]
grads_b = [grad_b]

mse = [np.mean((y - (a0*x + b0))**2)]
for k in range(100):
    a1 = a0 - alpha*grad_a
    b1 = b0 - alpha*grad_b
    
    a0 = a1
    b0 = b1
    
    grad_a = -2*np.mean( (y - a0*x - b0)*x)
    grad_b = -2*np.mean(y - a0*x - b0)
    
    aes.append(a1)
    bes.append(b1)
    grads_a.append(grad_a)
    grads_b.append(grad_b)
    
    error = np.mean((y - (a1*x + b1))**2)
    mse.append(error)
```

```python
plt.figure(figsize = (10,5))
plt.plot(mse)
plt.xlabel("IteraciÃ³n")
plt.ylabel("MSE")
plt.title("Mean Squared Error por iteraciÃ³n en el gradiente")
```

```python
_, axx = plt.subplots(nrows = 2, ncols = 2, figsize = (12,5))
axx[0,0].plot(aes)
axx[0,0].set_xlabel("Iteraciones")
axx[0,0].set_ylabel("Valor a")

axx[0,1].plot(bes)
axx[0,1].set_xlabel("Iteraciones")
axx[0,1].set_ylabel("Valor b")


axx[1,0].plot(grads_a)
axx[1,0].set_xlabel("Iteraciones")
axx[1,0].set_ylabel("Valor gradiente a")

axx[1,1].plot(grads_b)
axx[1,1].set_xlabel("Iteraciones")
axx[1,1].set_ylabel("Valor gradiente b")

pd.DataFrame({"a":aes, "grad_a": grads_a,
              "b":bes, "grad_b": grads_b})
```
